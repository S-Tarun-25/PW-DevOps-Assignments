#2. What is the difference between Hard Link & Soft Link?


**Hard Link**





*   It is a direct reference to the original file’s inode .

*   If the original file is deleted, the hard link still works because it points to the same data.

*   Cannot be created for directories .
*   Must be created within the same file system.


*   Both the original file and hard link share the same inode number.









```python
# Example to create:

ln file1 file2
```

**Soft Link**


*   It is like a shortcut to the original file,  it points to the file name, not the data.

*   If the original file is deleted, the soft link becomes broken.

*   Can be created for both files and directories.
*   Can link across different file systems.

*   The soft link has a different inode number from the original file.






```python
# Example to create:

ln -s file1 link1
```

#6. Step 1: Create user p1
 # Step 2: He should be part of 3 groups g1,g2,g3.
 # Step 3: whenever he creates a file automatically in the group section of file grp g1 should come.

```python
#Step 1: Create user p1:
sudo useradd p1

#Step 2: Add p1 to groups g1, g2, and g3:
sudo usermod -a -G g1,g2,g3 p1

#Step 3: Set the default group for p1 to g1:
sudo useradd -g g1 -G g2,g3 p1
```

#7. Step1: Create directory /tmp/bg as root user and create files inside it.

 # Step2: “abhi” should be the owner of the directory. He should be able to create files and delete files inside the directory and also he should be able to add content to all files inside the directory.

```python
#Step 1: As the root user, create the directory /tmp/bg:

sudo mkdir /tmp/bg

#Step 2: Change the ownership to "abhi" and set appropriate permissions:

sudo chown abhi:abhi /tmp/bg
sudo chmod 770 /tmp/bg
```
